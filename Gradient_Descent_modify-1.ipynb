**Examples:** Clustering, Dimensionality Reduction

**Methods:** K-Means, Hierarchical Clustering, PCA, etc.


##**Gradient Descent**
**Definition:** an optimization algorithm used to find the minimum of a function
How it works: starts with an initial guess, repeatedly updates the guess in the direction of the negative gradient (the direction of steepest descent) until the function reaches a local minimum

**Applications:** Linear Regression, Logistic Regression, Neural Networks

* In gradient descent, the cost function is a function that measures the error between the predicted and actual output of a machine learning model. The goal of gradient descent is to minimize this cost function by adjusting the model parameters iteratively.

* The cost function is typically defined by the specific problem and the type of model being used. For example, in linear regression, the cost function is often the mean squared error (MSE) between the predicted and actual outputs. In logistic regression, the cost function is often the cross-entropy loss.

* To compute the cost function, you first need to define it mathematically based on the problem and the model being used. Once the cost function is defined, you can evaluate it for a given set of model parameters and input data. 
* The cost function value is then used to calculate the gradient of the cost function with respect to the model parameters, which is then used to update the model parameters in the direction of decreasing cost in each iteration of the gradient descent algorithm.

Mean squared error (MSE) is a popular cost function used in regression problems to evaluate the performance of a machine learning model. It measures the average squared difference between the predicted values and the actual values.

The formula for MSE is:
   $$MSE = 1/n * (yi - i)^2$$

where:

* n is the number of data points
* yi is the actual value of the i-th data point
* i is the predicted value of the i-th data point

To compute the MSE, you need to:

Make predictions using your machine learning model for the given input data.
Calculate the difference between the predicted values and the actual values.
Square each difference.
Calculate the average of the squared differences.

For example, suppose you have a set of actual values [3, 6, 9, 12] and a set of predicted values [4, 5, 8, 11]. 

Then the MSE can be calculated as follows:

MSE = 1/4 * [(3-4)^2 + (6-5)^2 + (9-8)^2 + (12-11)^2]

= 1/4 * [1 + 1 + 1 + 1]

= 0.25 * 4

= 1

A lower MSE indicates better model performance, as it means that the predicted values are closer to the actual values.

```python
# Step 4: define the cost function (calculating the mean square error (MSE))
def cost_function(data, m, b):
    total_error = 0
    for i in range(len(data)):
        x = data[i, 0]

        y = data[i, 1] # actual value

        y_p = (m*x + b) # Find predicted value

        total_error += (y - y_p )**2  # find the difference in error square and summing them

    return total_error / float(len(data)) # finding the mean of the sum of error square
```

####**Step 5: perform the gradient descent**

The code bellow is an implementation of gradient descent algorithm to optimize the parameters m and b of a linear regression model.

Here's how it works:

1. The loop for i in range(1000) iterates 1000 times, which is the number of times the gradient descent algorithm will update the parameters **'m'** and **'b'**.

2. Inside the loop, **'m_grad'** and **'b_grad'** are initialized to 0. 
  * These variables will accumulate the gradients of the cost function with respect to m and b, respectively, for all data points.

3. **'N'** is the total number of data points in the dataset, and **for j in range(len(data))** iterates through each data point.

4. **'x'** and **'y'** are the feature and target variables for the j-th data point, respectively.

5. The gradient of the cost function with respect to **'m'** and **'b'** are computed using the current values of **'m'**, **'b'**, **'x'**, **'y'**, and **'N'**.

6. The updated values of m and b are computed by subtracting the learning rate times the gradients from their current values. The learning rate controls how big of a step the algorithm takes in each iteration.

Overall, this implementation of gradient descent algorithm performs the following steps for each iteration:

1. Compute the gradients of the cost function with respect to **'m'** and **'b'** for all data points.

2. Update the parameters **'m'** and **'b'** using the computed gradients and the learning rate.

This process is repeated for a fixed number of iterations or until the change in the cost function between iterations falls below a certain threshold. 

The end result is the optimal values of **'m'** and **'b'** that minimize the cost function and best fit the data.

```python
# Step 5: perform the gradient descent
for i in range(1000):
    m_grad = 0
    b_grad = 0
    N = float(len(data))
    for j in range(len(data)):
        x = data[j, 0]
        y = data[j, 1]
        m_grad += -(2/N) * x * (y - (m*x + b))
        b_grad += -(2/N) * (y - (m*x + b))
    m = m - (learning_rate * m_grad)
    b = b - (learning_rate * b_grad)
```

```python
# Step 6: print the final parameters
print("m:", m)
print("b:", b)
```

```python
# Let us generate a random regression dataset with 1 feature
X, y = make_regression(n_samples=100, n_features=1, noise=20)

# Initialize the coefficients
m = 0
b = 0

# Define the learning rate
alpha = 0.01

# Number of iterations
n_iter = 1000

# Perform gradient descent
for i in range(n_iter):
    y_pred = m * X + b
    error = y_pred - y
    m_gradient = 2/len(X) * np.dot(error, X)
    b_gradient = 2/len(X) * np.sum(error)
    m = m - alpha * m_gradient
    b = b - alpha * b_gradient

# Print the final coefficients
print("m:", m)
print("b:", b)
```

###**LogisticRegression**

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Lt us generate a random binary classification dataset
X, y = make_classification(n_samples=100, n_features=1, n_classes=2,
                           n_informative=1, n_redundant=0, n_repeated=0, n_clusters_per_class=1)

# Create and fit the Logistic Regression model
model = LogisticRegression()
model.fit(X, y)

# Print the coefficients
print("coef:", model.coef_)
print("intercept:", model.intercept_)
```

**Note**
* The learning rate (alpha) is too high: If the learning rate is set too high, the algorithm may overshoot the optimal solution and diverge.

* The number of iterations (n_iter) is too low: If the number of iterations is not sufficient, the algorithm may not have enough time to converge to the optimal solution.

* The initialization of the coefficients is not appropriate: If the initial coefficients are not set close enough to the optimal solution, the algorithm may take a long time to converge or may not converge at all.

* Data is not linearly separable: If the data is not linearly separable, no matter how many iteration you run, the algorithm will not be able to find the optimal solution.

* The cost function is not convex: If the cost function is not convex, the algorithm can get stuck in a local minimum.

* Logistic regression implemented with gradient descent is just an example and it is not meant to be used in real-world scenarios. 
* Libraries like scikit-learn have built-in logistic regression functions that are more efficient and better suited for real-world applications.
* Logistic regression is a sensitive model and therefore it is important to check the performance of the model using metrics such as accuracy, f1-score, precision, recall and AUC-ROC.

* Using the Gradient descent method to solve the linear regression and logistic regression problems is time consuming and computationally expensive, that's why libraries like scikit-learn use more efficient optimization algorithms like L-BFGS and others.
