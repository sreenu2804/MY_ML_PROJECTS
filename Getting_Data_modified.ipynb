##**Python File Open**

File handling is an important part of any web application.

Python has several functions for creating, reading, updating, and deleting files.


The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:

* "r" - Read - Default value. Opens a file for reading, error if the file does not exist

* "a" - Append - Opens a file for appending, creates the file if it does not exist

* "w" - Write - Opens a file for writing, creates the file if it does not exist

* "x" - Create - Creates the specified file, returns an error if the file exists

In addition you can specify if the file should be handled as binary or text mode

* "t" - Text - Default value. Text mode

* "b" - Binary - Binary mode (e.g. images)

````python
#To open a file for reading it is enough to specify the name of the file:

#Note: Make sure the file exists, or else you will get an error.

f = open("Myfile.txt") # To open the file called Myfile

f = open("Myfile.txt", "rt") #Because "r" for read, and "t" for text are the default values, you do not need to specify them.


print(f.read())

#Open a file on a different location:

#f = open("D:\\myfiles\welcome.txt", "r")
#print(f.read())
````

##**Downloadin Data from Google Trends**

Google Trends is a useful search trends feature that shows how frequently a given search term is entered into Google’s search engine relative to the site’s total search volume over a given period of time. Google Trends can be used for comparative keyword research and to discover event-triggered spikes in keyword search volume.

Google Trends provides keyword-related data including search volume index and geographical information about search engine users.

````python
# import the API
from pytrends.request import TrendReq # API for downloading google trends
pytrend = TrendReq()
````

````python
#We define a function that would allow us to pull this data for any state:
def get_searches1(key_word, state):
    pytrends = TrendReq(hl='en-US', tz=360)
    pytrends.build_payload(key_word, cat=0, timeframe='2020-01-01 2020-03-10',gprop='',geo=state)    
    df = pytrends.interest_over_time()
    return(df)
````

````python
#Build your keywords
keywrod = ["coronavirus news","coronavirus symptoms","what is coronavirus",
           "coronavirus","wuhan coronavirus"]

#declear your  country or Geo cation
# IT = Italy
dk = get_searches1(keywrod,"IT")

dk.head(30)
````
